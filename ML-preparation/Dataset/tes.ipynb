```python
import pandas as pd
import random

# Load Dataset from CSV
df = pd.read_csv('Harga-Rumah-Model.csv', delimiter=',')

# Mapping sub-lokasi berdasarkan kota
sub_lokasi_dict = {
    'Jakarta': ['Jakarta Utara', 'Jakarta Barat', 'Jakarta Selatan', 'Jakarta Timur', 'Jakarta Pusat'],
    'Bogor': ['Bogor Utara', 'Bogor Selatan', 'Bogor Timur', 'Bogor Barat', 'Bogor Tengah'],
    'Tangerang': ['Tangerang Kota', 'Tangerang Selatan', 'Tangerang Utara', 'Tangerang Barat'],
    'Depok': ['Depok'],
    'Bekasi': ['Bekasi Barat', 'Bekasi Timur', 'Bekasi Utara', 'Bekasi Selatan']
}

# Fungsi untuk mengisi nilai kolom 'KOTA' berdasarkan kondisi yang diberikan
def isi_kota(row):
    if pd.notna(row['JKT']) and pd.notna(row['GRS']):
        if row['JKT'] >= 5 and row['GRS'] == 'ADA':
            return random.choice(sub_lokasi_dict['Jakarta'])
        elif row['JKT'] < 5 and row['GRS'] == 'ADA':
            return random.choice(sub_lokasi_dict['Bogor'])
        elif row['JKT'] >= 5 and row['GRS'] != 'ADA':
            return random.choice(sub_lokasi_dict['Tangerang'])
        elif row['JKT'] < 5 and row['GRS'] != 'ADA':
            return random.choice(sub_lokasi_dict['Bekasi'])
    else:
        return row['KOTA']  # Jika ada nilai 'KOTA' yang sudah ada, gunakan nilai tersebut

# Mengisi nilai kolom 'KOTA' berdasarkan fungsi isi_kota
df['KOTA'] = df.apply(isi_kota, axis=1)

# Menyimpan DataFrame yang telah diubah ke dalam file CSV yang sama
df.to_csv('Harga-Rumah-Model.csv', sep=',', index=False)

# Informasi untuk user bahwa proses telah selesai
print("Data telah berhasil diubah dan disimpan kembali ke 'Harga-Rumah-Model.csv'.")
```
